```python
#Leemos el segundo archivo de los datos del suelo.
df1 = pd.DataFrame(pd.read_csv("soil-nogal-09.csv", engine="python"))
df1.head()
```

```python
#Checamos los tipos de variables:
df1.dtypes
```

```python
#Queremos únicamente las columnas: 2, 3, 4, Tsuelo_10cm, Tsuelo_20cm, Tsuelo_30cm, Tsuelo_40cm, Tsuelo_55cm, Tsuelo_70cm, 
#suelo_85cm y Tsuelo_100cm.
#Usamos el comando filter para ello:
df1 =  df1.filter(['2 Year_RTM  L','3 Day_RTM  L','4 Hour_Minute_RTM  L','Tsuelo_10cm','Tsuelo_20cm','Tsuelo_30cm','Tsuelo_40cm','Tsuelo_55cm','Tsuelo_70cm','Tsuelo_85cm','Tsuelo_100cm'], axis=1)
df1.head()
```

```python
#Para crear una columna de Fechas, se crearán dos columnas: minutos y horas.
#La columna 4 la convertimos a tipo string, es decir, caracteres, pues las horas están en formato de puros caracteres.
#Por ejemplo, las 15:30 están de la forma 1530.
df1['4 Hour_Minute_RTM  L'] = df1['4 Hour_Minute_RTM  L'].astype(str)
```

```python
#Llenaremos la columna de horas y de minutos.
#Declaramos los arreglo para las horas y los minutos aún sin dimensiones específicas:
HORA=[]
MINUTO=[]

#Usamos un contador i para que recorra todo el DataFrame df1
for i in range(0,len(df1)):
    #¿Tiene 4 caracteres?
    if (len(str(df1['4 Hour_Minute_RTM  L'][i]))==4):
        #¿La hora es 2400? Por el caso de cuando tenemos la hora 24:00, para cambiarla a 00:00
        if(str(df1['4 Hour_Minute_RTM  L'][i])[0:2]=='24'):
            HORA.append('00')
            MINUTO.append('00')
        else:
            HORA.append(str(df1['4 Hour_Minute_RTM  L'][i])[0:2])
            MINUTO.append(str(df1['4 Hour_Minute_RTM  L'][i])[2:4])
    #¿Tiene 3 caracteres (las primeras 9 horas)?
    elif (len(str(df1['4 Hour_Minute_RTM  L'][i]))==3):
        #Guardamos en HORA el primer caracter, y en MINUTO los otros dos:
        HORA.append(str(df1['4 Hour_Minute_RTM  L'][i])[0:1])
        MINUTO.append(str(df1['4 Hour_Minute_RTM  L'][i])[1:3])
    #¿Tiene 2 caracteres (Para los primeros 30 minutos del día)?
    elif (len(str(df1['4 Hour_Minute_RTM  L'][i]))==2):
        HORA.append('00')
        MINUTO.append(str(df1['4 Hour_Minute_RTM  L'][i])[0:2])
        
#Ahora ya podemos escribir el arreglo de los Días del DataFrame df1:
DÍAS = [df1['3 Day_RTM  L'][i] for i in range(0,len(df1))]
```

```python
#Crearemos un nuevo DataFrame con los días, horas y minutos del DataFrame df1
#La variable f guardará las columnas del nuevo DataFrame:
f = {'DÍAS': DÍAS, 'HORA': HORA, 'MINUTO': MINUTO}
df_dates = pd.DataFrame(data=f)
df_dates.head()
```

```python
#Checamos los tipos de variables del DataFrame df_dates:
df_dates.dtypes
```

```python
#Creamos en ese mismo DataFrame una columna DÍA, para aumentar un día a la hora 00:00.
#Establecemos un arreglo no definido para DÍA:
DÍA=[]

#Recorremos la variable n a lo largo del DataFrame df_dates.
for n in range(0,len(df_dates)):
    if (df_dates['HORA'][n]=='00' and df_dates['MINUTO'][n]=='00'):
        DÍA.append(df_dates['DÍAS'][n]+1)
    else:
        DÍA.append(df_dates['DÍAS'][n])
df_dates['DÍA']=DÍA
```

```python
#Mostramos de nuevo el DataFrame:
df_dates.head()
```

```python
#Ahora procedemos a crear una variable de tipo string para el Año, el número de día, hora y minuto.
#Declaramos un arreglos sin definir dates:
fechas=[]

#Corremos la variable i por todo el DataFrame df1:
for i in range(0,len(df1)):
    fechas.append('2009 ' + str(df_dates['DÍA'][i]) + ' ' + df_dates['HORA'][i] + ' ' + df_dates['MINUTO'][i])
```

```python
#Ahora, ese arreglo de Fechas lo pasamos a ser una variable tipo Datetime.
#Creamos el arreglo FECHA para ir guradando ahí las variables temporales.
FECHA=[]

#Recorremos la variable i en todo el arreglo df1:
for i in range(0,len(df1)):
    w = datetime.datetime.strptime(fechas[i],'%Y %j %H %M')
    F = w.isoformat(' ')
    FECHA.append(F)
```

```python
#Y ahora sí podemos agregar la columna FECHA al DataFrame df1:
df1['FECHAS'] = FECHA
df1.tail()
```

```python
#Introduciremos el concepto promedio móvil (rolling mean) como método de suavizar la evolución temporal de una serie de tiempo,
#y se pide reproducir las gráficas suavizadas.
#Agregamos de nuevo la variable FECHA:
df_10['FECHA'] = df2['FECHA']
df_20['FECHA'] = df2['FECHA']
df_40['FECHA'] = df2['FECHA']
df_85['FECHA'] = df2['FECHA']
df_graph2['FECHA'] = df2['FECHA']

#Creamos un nuevo DataFrame para los promedio móviles.
df_10roll = pd.DataFrame()
df_20roll = pd.DataFrame()
df_40roll = pd.DataFrame()
df_85roll = pd.DataFrame()
df_airTroll = pd.DataFrame()

#Ahora calculamos el promedio móvil de cada una de las temperaturas.
df_10roll["max10"] = df_10.Tsuelo_10cm_max.rolling(60, min_periods=1).mean()
df_10roll["min10"] = df_10.Tsuelo_10cm_min.rolling(60, min_periods=1).mean()
df_10roll["mean10"] = df_10.Tsuelo_10cm_mean.rolling(60, min_periods=1).mean()

df_20roll["max20"] = df_20.Tsuelo_20cm_max.rolling(60, min_periods=1).mean()
df_20roll["min20"] = df_20.Tsuelo_20cm_min.rolling(60, min_periods=1).mean()
df_20roll["mean20"] = df_20.Tsuelo_20cm_mean.rolling(60, min_periods=1).mean()

df_40roll["max40"] = df_40.Tsuelo_40cm_max.rolling(60, min_periods=1).mean()
df_40roll["min40"] = df_40.Tsuelo_40cm_min.rolling(60, min_periods=1).mean()
df_40roll["mean40"] = df_40.Tsuelo_40cm_mean.rolling(60, min_periods=1).mean()

df_85roll["max85"] = df_85.Tsuelo_85cm_max.rolling(60, min_periods=1).mean()
df_85roll["min85"] = df_85.Tsuelo_85cm_min.rolling(60, min_periods=1).mean()
df_85roll["mean85"] = df_85.Tsuelo_85cm_mean.rolling(60, min_periods=1).mean()

df_airTroll["max_air"] = df_graph2.airT_Avg_max.rolling(60, min_periods=1).mean()
df_airTroll["min_air"] = df_graph2.airT_Avg_min.rolling(60, min_periods=1).mean()
df_airTroll["mean_air"] = df_graph2.airT_Avg_mean.rolling(60, min_periods=1).mean()
```
